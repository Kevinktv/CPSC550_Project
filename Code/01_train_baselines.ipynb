# Train Baselines

Notebook-first workflow for grid search, best-config selection, and final baseline training on the `train` and `retrain` splits.

```python
from pathlib import Path

import json
import matplotlib.pyplot as plt

from notebook_workflows import (
    build_small_search_space,
    rank_search_results,
    results_to_dataframe,
    run_training_notebook_workflow,
)
```

```python
# Centralized config
DATASET = "cifar10"  # or "mufac"
RUN_TRAIN = True
RUN_RETRAIN = False
USE_WANDB = True
SMOKE_MODE = False
REUSE_EXISTING_TRAIN_GRID = True
SEED = 0
NUM_BANK_SEEDS = 3
EPOCHS = 25
DEVICE = "auto"
MOMENTUM = 0.9
IMAGE_SIZE = 32 if DATASET == "cifar10" else 224 # mufac images are 224x224
CLASS_WEIGHTING = "auto"

CHECKPOINT_DIR = Path("checkpoints")
RESULTS_DIR = Path("results") / "grid_search"
DATA_ROOT = None
TASK_MANIFEST = None
SAMPLES_CSV = None

SEARCH_SPACE = build_small_search_space(DATASET)
```

```python
training_outputs = run_training_notebook_workflow(
    dataset=DATASET,
    run_train=RUN_TRAIN,
    run_retrain=RUN_RETRAIN,
    search_space=SEARCH_SPACE,
    epochs=EPOCHS,
    seed=SEED,
    num_bank_seeds=NUM_BANK_SEEDS,
    use_wandb=USE_WANDB,
    smoke_mode=SMOKE_MODE,
    checkpoint_dir=CHECKPOINT_DIR,
    results_dir=RESULTS_DIR,
    data_root=DATA_ROOT,
    task_manifest=TASK_MANIFEST,
    samples_csv=SAMPLES_CSV,
    momentum=MOMENTUM,
    image_size=IMAGE_SIZE,
    class_weighting=CLASS_WEIGHTING,
    device=DEVICE,
    reuse_existing_train_grid=REUSE_EXISTING_TRAIN_GRID,
)

training_outputs
```

```python
for split_name, split_outputs in training_outputs["splits"].items():
    print(f"\n=== {split_name.upper()} SUMMARY ===")
    selection = split_outputs["grid_search"]
    if selection.get("selection_mode") == "reuse_train_best":
        print("Retrain reused the train-selected best configuration.")
        print("Source split:", selection["source_split"])
        print("Source best config:", selection["source_best_config_path"])
    else:
        if selection.get("loaded_from_cache"):
            print("Loaded existing train grid-search results from:", selection["results_dir"])
            print("Cache match mode:", selection.get("cache_match"))
        ranked = rank_search_results(selection["results"])
        display(results_to_dataframe(ranked))
    print("Best config:")
    print(json.dumps(selection["best_result"], indent=2))
    print("Canonical checkpoint:", split_outputs["bank_training"]["canonical_checkpoint"])
```

```python
for split_name, split_outputs in training_outputs["splits"].items():
    selection = split_outputs["grid_search"]
    if selection.get("selection_mode") == "reuse_train_best":
        print(f"Skipping validation-ranking plot for {split_name}: config reused from train.")
        continue
    ranked = rank_search_results(selection["results"])
    df = results_to_dataframe(ranked)
    if hasattr(df, "plot"):
        ax = df.plot(x="config_index", y="best_val_accuracy", kind="bar", title=f"{DATASET} {split_name} validation ranking")
        ax.set_ylabel("Best validation accuracy")
        plt.show()
```
