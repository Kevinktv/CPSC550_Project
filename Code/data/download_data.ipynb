```python
import os
from pathlib import Path
from huggingface_hub import snapshot_download
from torch.utils.data import DataLoader, Dataset
import torchvision
```

# CIFAR-10
https://docs.pytorch.org/vision/main/generated/torchvision.datasets.CIFAR10.html

```python
print("Downloading CIFAR-10...")
cifar_path = "./cifar-10"
os.makedirs(cifar_path, exist_ok=True)

train_dataset = torchvision.datasets.CIFAR10(
    root=cifar_path, 
    train=True, 
    download=True
)
print("CIFAR-10 ready.")
```

# MUFAC Dataset
https://huggingface.co/datasets/Dasool/MUFAC

[Optional] Huggingface Token

1. Log in or sign up at huggingface.co.

2. Go to your Settings > Access Tokens (or navigate directly to https://huggingface.co/settings/tokens).

3. Click New Token, give it a name (e.g., "unlearning-project"), and set the role to Read.

4. Copy the generated token string (it usually starts with hf_).

```python
# Optional: Uncomment and replace the string below with your actual token for Faster download.
# os.environ["HF_TOKEN"] = "REPLACE_WITH_YOUR_HUGGINGFACE_TOKEN"
```

```python
print("Downloading MUFAC from Hugging Face...")
mufac_path = "./mufac"
os.makedirs(mufac_path, exist_ok=True)

local_dir = snapshot_download(
    repo_id="Dasool/MUFAC", 
    repo_type="dataset",
    local_dir=mufac_path
)
print(f"MUFAC successfully downloaded to: {local_dir}")
```
